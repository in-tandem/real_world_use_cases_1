```python
train_data_path = 'all/my_training_data_can_be_used_as_is.csv'
test_data_path = 'all/my_test_data_categorized_can_be_used_as_is.csv'

train_data = panda.read_csv(train_data_path)
test_data = panda.read_csv(test_data_path)
train_data['max_cat_1'] = train_data.max_cat_1.apply(lambda x: 1 if x=='N' else 0)
```

```python
all_merged_data_path = 'all/yes_no_merged.csv'
all_merged_data = panda.read_csv(all_merged_data_path)
```

```python
## add the mean of months lag to test data and train data

mean_months_lag = all_merged_data.groupby(['card_id'])['month_lag'].mean()
```

```python
mean_months_lag = mean_months_lag.reset_index()
mean_months_lag.rename({'month_lag':'mean_month_lag'}, axis='columns', inplace = True)
mean_months_lag.head()
```

```python
## add most appearing merchant category _id

# most_appearing_merchant_cat_id = all_merged_data.groupby(['card_id','merchant_category_id']).count().sort_values('merchant_category_id')

# most_appearing_merchant_cat_id = most_appearing_merchant_cat_id.reset_index()
# # most_appearing_merchant_cat_id.drop_duplicates(subset = ['card_id'],keep='last', inplace = True)
# # most_appearing_merchant_cat_id.rename({'merchant_category_id':'most_mc_id'}, axis='columns', inplace = True)
# most_appearing_merchant_cat_id.head()
```

```python
most_appearing_merchant_cat_id = all_merged_data.groupby(['card_id','merchant_category_id']).count().reset_index().\
                sort_values('city_id').drop_duplicates(subset=['card_id'], keep='last')
# .reset_index(name='merchant_category_id').sort_values('merchant_category_id')

```

```python
# all_merged_data[all_merged_data['card_id']=='C_ID_415BB3A509'][['card_id', 'merchant_category_id']]

# most_appearing_merchant_cat_id = most_appearing_merchant_cat_id[['card_id', 'merchant_category_id']]
most_appearing_merchant_cat_id.rename({'merchant_category_id':'most_appearing_mc_id'}, inplace=True, axis='columns')
most_appearing_merchant_cat_id.head()
```

```python
# mean_months_added = panda.concat([all_merged_data, mean_months_lag], axis =1)
# mean_months_added.dropna(subset=['card_id'], inplace=True)

mean_months_added = all_merged_data.merge(mean_months_lag, how ='inner', on='card_id')

mean_months_added.head()
```
