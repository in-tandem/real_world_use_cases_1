```python
card_data_with_categories_and_installment = yes_no_data[['card_id','installments','category_1', 'category_2','category_3']]
```

```python
import timeit

mapper = {'A':3, 'B':2, 'C':1}

%timeit card_data_with_categories_and_installment['cat_3'] = card_data_with_categories_and_installment.category_3.apply(lambda x: mapper.get(x,0))
```

```python
card_data_with_categories_and_installment[['card_id','installments','category_1', 'category_2','cat_3']].to_csv('all/cat_used.csv')
```

```python
## mean installments grouped by card id

_mean_installments = card_data_with_categories_and_installment.groupby(['card_id'])['installments'].mean()

_max_category_1 = card_data_with_categories_and_installment.groupby(['card_id','category_1']).count().sort_values('category_1').drop_duplicates(subset=['card_id'],keep='last')

_max_category_2 = card_data_with_categories_and_installment.groupby(['card_id','category_2']).count().sort_values('category_2').drop_duplicates(keep='last')

_max_category_3 = card_data_with_categories_and_installment.groupby(['card_id','cat_3']).count().sort_values('cat_3').drop_duplicates(keep='last')


```
