```python
from sklearn.preprocessing import RobustScaler, StandardScaler, PowerTransformer
from sklearn.model_selection import KFold,train_test_split

import time
import lightgbm as lgb
import pandas as panda
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plot
```

```python
train_data = panda.read_csv('train_with_hist_new_with_merchant.csv')
test_data = panda.read_csv('test_with_hist_new_with_merchant.csv')
```

```python
train_data.head(1)
```

```python
y = train_data.target
x = train_data[[i for i in train_data.columns.tolist() if i not in ['card_id','target','merchant_id','merchant_group_id','merchant_id_y','merchant_id_x','merchant_id_x.1','category_1','category_4']]]
```

```python
sns.distplot(y)
plot.show()
```

```python

xtest = test_data[[i for i in test_data.columns.tolist() if i not in ['card_id','target','merchant_id','merchant_group_id','merchant_id_y','merchant_id_x','merchant_id_x.1','category_1','category_4']]]
```

```python
xtrain = x
ytrain = y
```

```python
from scipy import stats
def calculateCorrelationCoefficientsAndpValues(x_data, y_data, xlabel):
    
    pearson_coef, p_value = stats.pearsonr(x_data, y_data)
#     print("The Pearson Correlation Coefficient for %s is %s with a P-value of P = %s" %(xlabel,pearson_coef, p_value))
    
    return (pearson_coef,p_value)
```

```python
pearson_coeff = []
p_value = []
col_name = []

for col in [i for i in xtrain.columns.tolist() if i not in ['id','target']]:
    
    x,y = calculateCorrelationCoefficientsAndpValues(xtrain[col], train_data['target'], col)
    pearson_coeff.append(x)
    p_value.append(y)
    col_name.append(col)
    
pearson_table = panda.DataFrame({'column_name':col_name , 'pearson_coeff':pearson_coeff, 'p_value': p_value})
pearson_table.sort_values(by=['p_value'],ascending=True).head()
```

```python
reqd_cols = pearson_table.sort_values(by=['p_value'],ascending=True).head(25).column_name.values.tolist()
reqd_cols
```

```python
pearson_table[(pearson_table.pearson_coeff>0.05) | (pearson_table.pearson_coeff<-0.05)]
```

```python
xtrain_1 = xtrain[reqd_cols]
xtest_1 = xtest[reqd_cols]
```

```python
nfolds = 10
folds = KFold(n_splits= nfolds, shuffle=True, random_state=15)
param = {'num_leaves': 50,
         'min_data_in_leaf': 30, 
         'objective':'regression',
         'max_depth': 10,
         'learning_rate': 0.005,
         "min_child_samples": 100,
         "boosting": "gbdt",
         "feature_fraction": 0.9,
         "bagging_freq": 1,
         "bagging_fraction": 0.9 ,
         "bagging_seed": 11,
         'subsample' : 0.8,
         "metric": 'rmse',
         "lambda_l1": 0.1,
         "verbosity": -1,
#          "alpha":0.8
        
        }
```

```python
feature_importance_df = np.zeros((xtrain_1.shape[1], nfolds))
mvalid = np.zeros(len(xtrain))
mfull = np.zeros(len(xtest))


start = time.time()

# xtest = scaler.fit_transform(xtest)
# ytrain = pd.DataFrame(PowerTransformer().fit_transform(ytrain.values[:,np.newaxis]))
scaler = RobustScaler()
for fold_, (trn_idx, val_idx) in enumerate(folds.split(xtrain_1, ytrain)):
    print('----')
    print("fold nÂ°{}".format(fold_))
    
    x0,y0 = xtrain_1.iloc[trn_idx], ytrain[trn_idx]
    x1,y1 = xtrain_1.iloc[val_idx], ytrain[val_idx]
    
#     x0 = panda.DataFrame(scaler.fit_transform(x0))
#     x1 = panda.DataFrame(scaler.transform(x1))
#     y0 = panda.DataFrame(scaler.fit_transform(y0.values.reshape(-1,1)))
#     y1 = panda.DataFrame(scaler.fit_transform(y1.values.reshape(-1,1)))
    
    trn_data = lgb.Dataset(x0, label= y0); val_data = lgb.Dataset(x1, label= y1)
    
    num_round = 10000
    clf = lgb.train(param, trn_data, num_round, valid_sets = [trn_data, val_data], 
                    verbose_eval=500, early_stopping_rounds = 150)
    mvalid[val_idx] = clf.predict(x1, num_iteration=clf.best_iteration)
    
    feature_importance_df[:, fold_] = clf.feature_importance()
    
    mfull += clf.predict(xtest_1, num_iteration=clf.best_iteration) / folds.n_splits
```

```python
clf.feature_importance().shape
# list(zip(clf.feature_importance(),xtrain.columns))
```

```python
feature_imp = panda.DataFrame(sorted(zip(clf.feature_importance(),xtrain.columns),reverse=True), columns=['value','feature'])
imp_columns = feature_imp.loc[feature_imp.value>1000].feature.values.tolist()
imp_columns
```

```python
x_1 = xtrain[imp_columns]
x_test_1 = xtest[imp_columns]
```

```python
ximp = panda.DataFrame()
ximp['feature'] = xtrain.columns
ximp['importance'] = feature_importance_df.mean(axis = 1)

plot.figure(figsize=(14,14))
sns.barplot(x="importance",
            y="feature",
            data=ximp.sort_values(by="importance",
                                           ascending=False))
plot.title('LightGBM Features (avg over folds)')
plot.tight_layout()
```

```python
final = panda.DataFrame(mfull)
final.columns = ['target']

final.head()
```
