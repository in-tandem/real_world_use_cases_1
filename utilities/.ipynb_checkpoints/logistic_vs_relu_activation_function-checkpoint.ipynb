### Logistic Sigmoid Activation Function

This particular function is used as an activation function to keep smaller numbers close to 0 and larger numbers close to 1. Output values for the logistic sigmoid function will never exceed 0 or 1.

This particular function is non-linear. In the sense, values will be in the range of 0-1 and not 0/1.

Logistic sigoid function is also fully differentiable, making it an attractive option for use as an activation function in deep neural networks. Lets take a look at the following example to understand logistic sigmoid function better.

Lets take the example of y = f(x), where f is the logistic sigmoid function

to read this:

<li> y is the probability of a neuron firing. Since its a probability , it ranges within 0-1</li>
<li> x is the amount of evidence for a certain feature. Based on which probability of neuron firing will be calculated</li>

##### Observations we can make are:

1. When x =0, y reached a value of 0.5
2. When x is lower than zero, y gravitated towards value of 0
3. When x is higher than zero, y gravitated towards value of 1
4. When it comes to derivative, derivative is highest when x = 0
5. For values lower/higher than x = 0, derivative gradually descends to 0


What is the meaning of point 5? Lets see with an example.

Lets take the example of score vs effort put in. The figure b, shows change in scores on y axis and effort put in x axis. 

The amount of effort required to get test score of 80 vs test score of 40 is quite high. So change in score(slope of y/slope in figure b) is steeper with respect to effort put in.

Irrespective of the effort put in, the change in scores from 80 to say 84 will be slower. Or to put it in another way, after a certain point in score(say threshold of 80) , the change is score with respect to effort put in will be slower, i.e less steeper. As shown in right side of the peak in figure b.

Along similar arguments, change in scores will be slower with gradually increasing efforts.
